### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:

*   оценка 2 и ниже — низкий рейтинг;
*   оценка 4 и ниже — средний рейтинг;
*   оценка 4.5 и 5 — высокий рейтинг.

Результат классификации запишите в столбец class.

````python
df_ratings = pd.read_csv('ratings.csv', sep=',', encoding='utf-8')
df_ratings.head()
````

````python
df_movies = pd.read_csv('movies.csv', sep=',', encoding='utf-8')
df_movies.head()
````

````python
def categorize_movies_by_grade(rating):
    """
    Функция позволяет классифицировать фильмы по правилам:
    * оценка 2 и ниже — низкий рейтинг;
    * оценка 4 и ниже — средний рейтинг;
    * оценка 4.5 и 5 — высокий рейтинг.
    """
    if rating <= 2:
        return 'низкий рейтинг'
    elif 2 < rating <= 4:
        return 'средний рейтинг'
    elif 4.5 <= rating <= 5:
        return 'высокий рейтинг'
````

````python
df_ratings['class'] = df_ratings['rating'].apply(categorize_movies_by_grade)
df_ratings
````

### Задание 2
Используйте файл keywords.csv.

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
```
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```
Результат классификации запишите в отдельный столбец region.

````python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}


df_keywords = pd.read_csv('keywords.csv', sep=',', encoding='utf-8')
df_keywords
````

````python
def exhibit_geographic_affiliation(query_string_list):
    """
    Функция позволяет создать гео-классификатор, который каждой строке сможет 
    выставить географическую принадлежность определённому региону. Т.е. если 
    поисковый запрос содержит название города региона, то в столбце 'region' 
    пишется название этого региона. Если поисковый запрос не содержит названия 
    города, то ставим 'undefined'.
    """
    for region, list_city in geo_data.items():
        for city in list_city:
            if city in query_string_list:
                return region
    return 'undefined'
````

````python
# записываем результат в столбец region
df_keywords['region'] = df_keywords['keyword'].apply(exhibit_geographic_affiliation)
df_keywords
````

### Задание 3 (бонусное)

Есть мнение, что раньше снимали настоящее кино, не то что сейчас. Ваша задача — проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия: файл movies.csv и ratings.csv из базы. Нужно проверить, верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

Вы не будете затрагивать субьективные факторы выставления этих рейтингов, а пройдётесь по алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010 года.
2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте алгоритм:

    * для каждой строки пройдите по всем годам списка years;
    * если номер года присутствует в названии фильма, то функция возвращает этот год, как год выпуска;
    * если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год.
3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’.
4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга.

````python
years = [] # создание списка для дальнейшего заполнения его годами
start_date = 1950
end_date = 2010

# наполняется список года между 1950 и 2010
while start_date < end_date:
    start_date = start_date + 1
    years.append(str(start_date))
````

````python
def production_year(df):
    """
    Функция позволяет каждой строке из названия фильма выставлять год выпуска.
    Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому 
    используется алгоритм:  
    * для каждой строки пройдите по всем годам списка years;
    * если номер года присутствует в названии фильма, то функция возвращает этот 
        год, как год выпуска;
    * если ни один из номеров года списка years не встретился в названии фильма, 
        то возвращается 1900 год.
    """
    for year in years: 
        if year in df:
            return year
    return '1990'

df_movies['year'] = df_movies['title'].apply(production_year)
df_movies
````

````python
# объединяем таблицы для дальнейшего посчета среднего по оценке
df_ratings_movies = df_ratings.merge(df_movies, left_on='movieId', right_on='movieId', how='inner')[['movieId', 'rating', 'title', 'class', 'year']]
df_ratings_movies
````

````python
# подсчет среднего значения, затем сортируем результат средней оценки по убыванию
result = df_ratings_movies[['rating', 'title']].groupby(df_ratings_movies['title']).mean().reset_index()
result.sort_values(by='rating', ascending=False)
````
