### Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
df_rating = pd.read_csv('ratings.csv')
df_rating.head()
```

```python
""" 
преобразовываем числовое значение timestamp в datetime и записываем в новый 
атрибут 
"""
df_rating['date'] = pd.to_datetime(df_rating['timestamp'], unit='s')
df_rating.head()
```

```python
"""
результат вычисления calculation_result: количество оценок с столбце 'rating' 
по каждому пользователю, max и min столбца 'timestamp'. 
"""
calculation_result = df_rating.groupby('userId').agg({'rating': ['count'], 'date': ['max', 'min']}).reset_index()[['userId', 'date', 'rating']]
calculation_result.head()
```

```python
# получаем данные где количество оценок пользовтаеля больше 100
filter_rating = calculation_result.loc[(calculation_result['rating']['count'] > 100)]
filter_rating.head()
```

```python
# получаем время жизни пользователей
filter_rating['result'] = (filter_rating['date']['max'] - filter_rating['date']['min']).dt.days
filter_rating.head()
```

```python
# получаем среднее время жизни пользователей
average_lifetime = filter_rating.result.mean()
average_lifetime
```

### Задание 4
Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе «Материалы для лекции “Продвинутый pandas”» ---- Ноутбуки к лекции «Продвинутый pandas»).
Нужно сформировать две таблицы:

* таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
* аналогичную таблицу по типам выручки с указанием адреса клиента.
Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

```python
# rzd - железнодорожные перевозки
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

# auto - автомобильные перевозки
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

# air - воздушные перевозки
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

# client_base - адреса клиентов
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
"""
получаем все данные с помощью соединения full outer join таблиц air, auto, air
"""
result_rzd_auto = pd.merge(air, auto, how='outer', on='client_id')
result_rzd_auto_air = pd.merge(result_rzd_auto, air, how='outer', on='client_id')
result_rzd_auto_air
```

```python
"""
результат соединения таблицы result_rzd_auto_air с client_base для получения 
адресов с помощью соединения full outer join
"""
result_df = pd.merge(result_rzd_auto_air, client_base, how='outer', on='client_id')
result_df
```
