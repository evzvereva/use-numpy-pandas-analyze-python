### Задание 1
Скачайте с [сайта](https://grouplens.org/datasets/movielens/) датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
df_film = pd.read_csv('movies.csv', sep=',')
df_film.head()
```

```python
df_ratings = pd.read_csv('ratings.csv', sep=',')
df_ratings.head()
```

```python
df_merge_ratings_film = df_film.merge(df_ratings, how="inner", on="movieId") # соединяем датафреймы по атрибуту "movieId"
columns_movieid_rating = df_merge_ratings_film[["movieId", "rating"]] # получаем данные по атрибута id фильма и рейтинга
columns_movieid_rating.head()
```

```python
# подсчитываем количество рейтинга по каждому фильму, где рейтинг = 5.0
count_rating = columns_movieid_rating[columns_movieid_rating["rating"] == 5.0].groupby(["movieId"]).count().reset_index()
count_rating.sort_values(by=["rating"], ascending=False)
```

```python
# находим максимальное количество оценок по id фильма
max_rating = count_rating[count_rating["rating"] == count_rating["rating"].max()]
max_rating.head()
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
df_power = pd.read_csv('power.csv', sep=',')
```

```python
df_power.info()
```

```python
"""
получаем данные за период с 2005 по 2010 включительно где quantity не 
должен быть отрицательным, а также по category, которая содержит 4, 12, 21
"""
result = df_power.query('year >= 2005 and year <= 2010 and quantity >= 0')
result_category = result[result['category'].isin([4, 12, 21])]
result_country = result_category[result_category['country'].isin(['Latvia', 'Lithuania', 'Estonia'])]
result_country
```
